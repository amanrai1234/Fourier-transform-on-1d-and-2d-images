```python
#disscrete fourier trasnform using two 1d fourier transforms
def dft_matrix(image):
    r,c = image.shape
    t = np.zeros((r,c),complex)
    transformed_img = np.zeros((r,c),complex)
    w = np.arange(r)
    x = np.arange(c)
    y = w.reshape((r,1))
    z = x.reshape((c,1))
    for a in range(0,r):
        M1 = 1j*np.sin(-2*np.pi*z*x/c) + np.cos(-2*np.pi*z*x/c)
        t[a] = np.dot(M1, image[a])
    for b in range(0,c):
        M2 = 1j*np.sin(-2*np.pi*y*w/c) + np.cos(-2*np.pi*y*w/c)
        transformed_img[:,b] = np.dot(M2, t[:,b])
    return transformed_img
```
