```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
N=int(input("enter the number of samples in between the two numbers"))
#N=100
a=int(input("enter the llowe limit value of the function"))
b=int(input("enter the upper limit value of the function"))

x=np.linspace(a,b, N)
y=np.zeros(N)
#this is the test input function
y[abs(x)< 1] = x[abs(x) < 1]
plt.plot(x,y)
plt.show()
```

```python
c=np.zeros(N, complex)
n=np.arange(N)
for k in range(N):
    c[k] =np.sum(y*np.exp(-2j*np.pi*k*n/N))
plt.figure(figsize=(12,4))    
plt.subplot(1,2,1)


plt.plot(np.imag(c))

plt.plot(np.real(c))
plt.subplot(1,2,2)
plt.plot(abs(c))
plt.legend()
plt.show()
```

```python
#inverse

y2=np.zeros(N, complex)
k=np.arange(N)
for n in range(N):
    y2[n]=1/N*np.sum(c*np.exp(2j*np.pi*k*n/N))

plt.plot(np.real(y2))
plt.show()
```
